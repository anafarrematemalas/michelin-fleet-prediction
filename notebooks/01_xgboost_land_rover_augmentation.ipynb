### GROUPED BY COUNTRY AND BRAND AND MODEL

````python
df_grouped = (
    train
    .groupby(["country_name", "car_maker_name", "car_segment_name"], as_index=False)
    ["total_vehicles"]
    .sum()
)
````

````python
df_grouped
````

### ADDING THE MOST SIMILLAR COUNTRY PER ROW OF TURKEY

````python
df_turkey = df_grouped[df_grouped["country_name"] == "Turkey"].copy()
df_others = df_grouped[df_grouped["country_name"] != "Turkey"].copy()
````

````python
df_compare = df_turkey.merge(
    df_others,
    on=["car_maker_name", "car_segment_name"],
    suffixes=("_turkey", "_other")
)
````

````python
df_compare["diff"] = (
    df_compare["total_vehicles_turkey"]
    - df_compare["total_vehicles_other"]
).abs()
````

````python
df_closest = (
    df_compare
    .sort_values("diff")
    .groupby(["car_maker_name", "car_segment_name"])
    .first()
    .reset_index()
)
````

````python
df_closest.head(50)
````

````python
df_closest["country_name_other"].value_counts()
````

### FEATURE ENGINEERING

````python
# Diccionario para lookup rápido
closest_country_dict = df_closest.set_index(
    ["car_maker_name", "car_segment_name"]
)["country_name_other"].to_dict()

# Función para asignar peso
def get_weight(row):
    key = (row["car_maker_name"], row["car_segment_name"])
    closest_country = closest_country_dict.get(key, None)
    return 1.0 if row["country_name"] == closest_country else 0.3

train["weight"] = train.apply(get_weight, axis=1)
````

````python
train
````

````python
from sklearn.preprocessing import OneHotEncoder
import numpy as np

features = ["country_name", "car_maker_name", "car_segment_name"]
target = "total_vehicles"

# One-hot encoding
ohe = OneHotEncoder(sparse_output=False, handle_unknown="ignore")
X_train = ohe.fit_transform(train[features])

# 🔥 LOG TRANSFORM AQUÍ
y_train = np.log1p(train[target].values)

sample_weight = train["weight"].values

# Test set
X_test = ohe.transform(test[features])
````

````python
X_train
````

````python
y_train
````

````python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
import xgboost as xgb

models = {
    "RandomForest": RandomForestRegressor(n_estimators=200, max_depth=10, random_state=42, n_jobs=-1),
    "GradientBoosting": GradientBoostingRegressor(n_estimators=200, max_depth=5, random_state=42),
    "XGBoost": xgb.XGBRegressor(n_estimators=200, max_depth=5, random_state=42, n_jobs=-1),
    "LinearRegression": LinearRegression()
}
````

````python
results = []

def wmse(y_true, y_pred):
    weights = y_true
    return np.sum(weights * (y_true - y_pred)**2) / np.sum(weights)

def wmae(y_true, y_pred):
    weights = y_true
    return np.sum(weights * np.abs(y_true - y_pred)) / np.sum(weights)

for name, model in models.items():

    if name == "LinearRegression":
        model.fit(X_train, y_train)
    else:
        model.fit(X_train, y_train) #, sample_weight=sample_weight)

    preds_log = model.predict(X_test)
    preds = np.expm1(preds_log)

    wmse_model = wmse(test["total_vehicles"].values, preds)
    wmae_model = wmae(test["total_vehicles"].values, preds)
    wmse_baseline = wmse(
        test["total_vehicles"].values,
        test["baseline_total_vehicles"].values
    )

    improvement = wmse_baseline - wmse_model

    results.append({
        "model": name,
        "WMSE_model": wmse_model,
        "WMAE_model": wmae_model,
        "WMSE_baseline": wmse_baseline,
        "Improvement_over_baseline": improvement
    })

results_df = pd.DataFrame(results).sort_values(
    "Improvement_over_baseline",
    ascending=False
)

results_df
````

````python
import numpy as np

best_model_name = results_df.iloc[0]["model"]
best_model = models[best_model_name]

# Predicción en log
preds_log = best_model.predict(X_test)

# Volver a escala original
test["pred_total_vehicles"] = np.expm1(preds_log)

test[[
    "country_name",
    "car_maker_name",
    "car_segment_name",
    "total_vehicles",
    "baseline_total_vehicles",
    "pred_total_vehicles"
]].head(20)
````

# MEJORAR MODELO

````python
import numpy as np
import pandas as pd

age_mapping = {
    "Less than 1 year old": 0.5,
    "1 year old": 1,
    "2 years old": 2,
    "3 to 5 years old": 4,
    "6 to 10 years old": 8,
    "11 years and older": 12
}

train["code_age_num"] = train["code_age"].map(age_mapping)
test["code_age_num"] = test["code_age"].map(age_mapping)

train["code_age_num"] = train["code_age_num"].fillna(0)
test["code_age_num"] = test["code_age_num"].fillna(0)
````

````python
combo_cols = [
    "car_maker_name",
    "car_segment_name",
    "car_type_name",
    "energy",
    "code_age",
    "body_style"
]

# Total por país
country_totals = (
    train.groupby("country_name")["total_vehicles"]
    .sum()
    .reset_index()
    .rename(columns={"total_vehicles": "country_total"})
)

train = train.merge(country_totals, on="country_name")

# Share dentro del país
train["share"] = train["total_vehicles"] / train["country_total"]

# Share promedio por combinación
share_mean = (
    train.groupby(combo_cols)["share"]
    .mean()
    .reset_index()
    .rename(columns={"share": "mean_share"})
)

train = train.merge(share_mean, on=combo_cols, how="left")

# Baseline train
train["baseline_total_vehicles"] = (
    train["mean_share"] * train["country_total"]
)
````

````python
train["ratio"] = train["total_vehicles"] / train["baseline_total_vehicles"]

train["ratio"] = (
    train["ratio"]
    .replace([np.inf, -np.inf], 0)
    .fillna(0)
)

train["ratio"] = train["ratio"].clip(0, 5)  # evita ratios absurdos

y_train = np.log1p(train["ratio"].values)
````

````python
features_cat = [
    "country_name",
    "car_maker_name",
    "car_segment_name",
    "car_type_name",
    "energy",
    "body_style"
]

features_num = ["code_age_num", "country_total"]
````

````python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(
    sparse_output=False,
    handle_unknown="ignore"
)

# Ensure 'country_total' is merged into test before feature creation
# country_totals was already created from the train set in a previous cell
test = test.merge(country_totals, on="country_name", how="left")

X_train_cat = ohe.fit_transform(train[features_cat])
X_test_cat = ohe.transform(test[features_cat])

X_train = np.hstack([X_train_cat, train[features_num].values])
X_test = np.hstack([X_test_cat, test[features_num].values])
````

````python
import xgboost as xgb

sample_weight = train["total_vehicles"].values

model = xgb.XGBRegressor(
    n_estimators=900,
    max_depth=5,
    learning_rate=0.03,
    subsample=0.85,
    colsample_bytree=0.85,
    random_state=42,
    n_jobs=-1
)

model.fit(
    X_train,
    y_train,
    sample_weight=sample_weight
)
````

````python
ratio_log_pred = model.predict(X_test)

ratio_pred = np.expm1(ratio_log_pred)
ratio_pred = np.clip(ratio_pred, 0, None)

test["pred_total_vehicles"] = (
    test["baseline_total_vehicles"] * ratio_pred
)
````

````python
def wmse(y_true, y_pred):
    weights = y_true
    return np.sum(weights * (y_true - y_pred)**2) / np.sum(weights)

wmse_model = wmse(
    test["total_vehicles"].values,
    test["pred_total_vehicles"].values
)

wmse_baseline = wmse(
    test["total_vehicles"].values,
    test["baseline_total_vehicles"].values
)

print("WMSE Model:", wmse_model)
print("WMSE Baseline:", wmse_baseline)
print("Improvement:", wmse_baseline - wmse_model)
````

# Notebook Summary — `01_xgboost_land_rover_augmentation.ipynb`

---

## Purpose

Predict `total_vehicles` for 254 Turkey test configurations using XGBoost with two core innovations:
a log-ratio target that learns to correct the provided baseline, and synthetic data augmentation
to solve the cold-start problem for Land Rover configurations entirely absent from Turkey's training data.

---

## Pipeline (step by step)

### 1. Data loading
- Train: 91,763 rows × 10 columns — 11 European countries
- Test: 254 rows — Turkey only
- After cleaning (dropna): 91,697 train rows, 251 test rows

### 2. Cold-start discovery
Land Rover has **0 rows in Turkey's training data** — the model has never seen this brand in Turkey:
```python
land_rover_turkey = train[(train["car_maker_name"] == "LAND ROVER") &
                          (train["country_name"] == "Turkey")]
# → Empty DataFrame (0 rows)
```
Without augmentation, XGBoost predicts ~60 vehicles for Land Rover configs (actual: 18–96).

### 3. First model attempt (naive log1p target)
- Features: country, maker, segment (one-hot encoded)
- Target: `log1p(total_vehicles)`
- All 4 models tested (XGBoost, RandomForest, GradientBoosting, LinearRegression) **worse than baseline**
- Best: XGBoost WMSE = 1.35B vs baseline 1.005B

**Why it failed**: predicting raw vehicle counts directly doesn't account for the fact that a 50,000-vehicle
config and a 100-vehicle config are on completely different scales. The model learns an average that satisfies
neither.

### 4. Improved model — log-ratio target + vehicle weighting

Three changes transform the approach:

| Change | Why it helps |
|--------|-------------|
| Target: `log1p(total_vehicles / baseline_total_vehicles)` | Learn to *correct* the baseline, not predict from scratch. Baseline already encodes market-size signal. |
| `sample_weight = total_vehicles` | High-volume configs dominate training loss — aligned with WMSE metric. |
| Richer features: `code_age_num`, `country_total` | Adds numerical age and market size as continuous signals. |

```python
y_train = np.log1p(train["ratio"])           # ratio = total / baseline
model.fit(X_train, y_train, sample_weight=train["total_vehicles"])
pred = np.expm1(model.predict(X_test)) * test["baseline_total_vehicles"]
```

**Result after this change: WMSE = 554M** (baseline 1,005M → −45%)

### 5. Land Rover synthetic augmentation

Analysed Land Rover across 10 other countries (1,225 rows):
- Top segments: Range Rover Evoque, Defender, Range Rover Sport
- Body style: 4X4 (72%), SUV (28%)
- Energy: DIES (37%), ESS (29%), DIES-MHEV (11%)

Generated **50 synthetic Turkey Land Rover rows** by sampling features and `total_vehicles`
from other countries' LR distribution:

```python
synthetic_data = land_rover_other_countries.sample(n=50, replace=True, random_state=42)
synthetic_data["country_name"] = "Turkey"
synthetic_data["total_vehicles"] = land_rover_other_countries["total_vehicles"].sample(n=50)
train = pd.concat([train, synthetic_data])
```

Retrained XGBoost with augmented dataset (91,747 rows).

**Final result: WMSE = 356M** (baseline 920M → **−61%**)
